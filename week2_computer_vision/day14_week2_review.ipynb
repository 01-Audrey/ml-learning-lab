```python
# ============================================================
# SECTION 2: PERFORMANCE EVOLUTION
# ============================================================

print("\n" + "="*60)
print("PERFORMANCE EVOLUTION ANALYSIS")
print("="*60)

# Version data
versions = ['v1\nBaseline', 'v2\nOptimized', 'v3\nProduction']
mAP_scores = [15.8, 52.6, 75.1]
precision = [3.8, 79.8, 73.5]
recall = [31.5, 46.7, 72.1]
dataset_sizes = [66, 66, 246]
epochs = [10, 50, 100]

print("""
VERSION COMPARISON:
==================

v1 (BASELINE) - Day 10:
  ‚Üí mAP@50: 15.8%
  ‚Üí Dataset: 66 images
  ‚Üí Model: YOLOv8n
  ‚Üí Epochs: 10
  ‚Üí Training time: ~1.5 minutes
  ‚Üí Status: Established baseline ‚ùå

v2 (OPTIMIZED) - Day 11:
  ‚Üí mAP@50: 52.6% (+233% improvement!)
  ‚Üí Dataset: 66 images (same)
  ‚Üí Model: YOLOv8n (same)
  ‚Üí Epochs: 50 (‚Üë 5x)
  ‚Üí Training time: ~10 minutes
  ‚Üí Status: Validated training approach ‚ö°

v3 (PRODUCTION) - Day 12:
  ‚Üí mAP@50: 75.1% (+375% from v1!) ‚úÖ
  ‚Üí Dataset: 246 images (‚Üë 3.7x)
  ‚Üí Model: YOLOv8s (‚Üë 3.4x capacity)
  ‚Üí Epochs: 100 (‚Üë 2x from v2)
  ‚Üí Training time: ~35 minutes
  ‚Üí Status: PRODUCTION READY! üöÄ
""")

# Calculate improvements
v1_to_v2_improvement = ((mAP_scores[1] - mAP_scores[0]) / mAP_scores[0]) * 100
v2_to_v3_improvement = ((mAP_scores[2] - mAP_scores[1]) / mAP_scores[1]) * 100
total_improvement = ((mAP_scores[2] - mAP_scores[0]) / mAP_scores[0]) * 100

print(f"""
IMPROVEMENT ANALYSIS:
====================
v1 ‚Üí v2: +{v1_to_v2_improvement:.1f}% improvement
v2 ‚Üí v3: +{v2_to_v3_improvement:.1f}% improvement
v1 ‚Üí v3: +{total_improvement:.1f}% TOTAL improvement

TARGET: 70% mAP
ACHIEVED: 75.1% mAP
EXCEEDED TARGET BY: +5.1%
""")
```

```python
# ============================================================
# SECTION 3: VISUALIZATION - PERFORMANCE EVOLUTION
# ============================================================

print("\n" + "="*60)
print("CREATING PERFORMANCE VISUALIZATIONS")
print("="*60)

# Create figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Week 2: Safety Equipment Detector - Performance Evolution', 
             fontsize=16, fontweight='bold', y=0.995)

# Plot 1: mAP Evolution
ax1 = axes[0, 0]
bars1 = ax1.bar(versions, mAP_scores, color=['#ff6b6b', '#feca57', '#1dd1a1'], 
                edgecolor='black', linewidth=1.5)
ax1.axhline(y=70, color='red', linestyle='--', linewidth=2, label='Target: 70%', alpha=0.7)
ax1.set_ylabel('mAP@50 (%)', fontsize=11, fontweight='bold')
ax1.set_title('mAP@50 Evolution', fontsize=12, fontweight='bold')
ax1.set_ylim(0, 100)
ax1.legend()
ax1.grid(axis='y', alpha=0.3)

# Add value labels on bars
for bar, score in zip(bars1, mAP_scores):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
             f'{score:.1f}%', ha='center', va='bottom', fontweight='bold')

# Plot 2: Precision vs Recall
ax2 = axes[0, 1]
x = np.arange(len(versions))
width = 0.35
bars2a = ax2.bar(x - width/2, precision, width, label='Precision', 
                 color='#4834d4', edgecolor='black')
bars2b = ax2.bar(x + width/2, recall, width, label='Recall', 
                 color='#6c5ce7', edgecolor='black')
ax2.set_ylabel('Score (%)', fontsize=11, fontweight='bold')
ax2.set_title('Precision & Recall Evolution', fontsize=12, fontweight='bold')
ax2.set_xticks(x)
ax2.set_xticklabels(versions)
ax2.legend()
ax2.set_ylim(0, 100)
ax2.grid(axis='y', alpha=0.3)

# Plot 3: Dataset Growth
ax3 = axes[1, 0]
bars3 = ax3.bar(versions, dataset_sizes, color=['#ff6b6b', '#feca57', '#1dd1a1'],
                edgecolor='black', linewidth=1.5)
ax3.set_ylabel('Number of Images', fontsize=11, fontweight='bold')
ax3.set_title('Dataset Size Growth', fontsize=12, fontweight='bold')
ax3.grid(axis='y', alpha=0.3)

for bar, size in zip(bars3, dataset_sizes):
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height,
             f'{size}', ha='center', va='bottom', fontweight='bold')

# Plot 4: Training Configuration
ax4 = axes[1, 1]
bars4 = ax4.bar(versions, epochs, color=['#ff6b6b', '#feca57', '#1dd1a1'],
                edgecolor='black', linewidth=1.5)
ax4.set_ylabel('Number of Epochs', fontsize=11, fontweight='bold')
ax4.set_title('Training Duration', fontsize=12, fontweight='bold')
ax4.grid(axis='y', alpha=0.3)

for bar, epoch in zip(bars4, epochs):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height,
             f'{epoch}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.savefig('../results/week2_performance_evolution.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n‚úÖ Performance visualizations saved!")
print("   ‚Üí File: results/week2_performance_evolution.png")

```
